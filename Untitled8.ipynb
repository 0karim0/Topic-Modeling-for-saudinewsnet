# **Import libraries**

```python
from datasets import load_dataset

import pandas as pd

import re
import random

import string
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

from sentence_transformers import SentenceTransformer

from umap import UMAP

from hdbscan import HDBSCAN

from sklearn.feature_extraction.text import CountVectorizer
import arabicstopwords.arabicstopwords as stp


import matplotlib.pyplot as plt
import seaborn as sns

from datetime import datetime
```

```python
df = load_dataset("saudinewsnet")
```

```python
df['train'][0]
```

```python
df = [
    {
        "text": rec["content"],
        "source": rec["source"],
        "date": rec["date_extracted"]
    }
    for rec in df['train']
]
```

```python
df = pd.DataFrame(df)
# shuffle data
df = df.sample(frac=1, random_state=101)
```

```python
df.head()
```

```python
df.shape
```

# Text Data Preprocessing

```python
nltk.download('punkt')

def clean_text(text):
    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove English letters
    text = re.sub(r'[a-zA-Z]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text)

    # Tokenize text
    tokens = word_tokenize(text)

    # Remove stop words
    arabic_stopwords = set(stopwords.words('arabic'))
    tokens = [word for word in tokens if word not in arabic_stopwords]

    # Normalize text (optional)
    # For example, you can apply stemming or lemmatization

    # Join tokens back into a single string
    preprocessed_text = ' '.join(tokens)

    return preprocessed_text

```

```python
df['text'] = df['text'].apply(clean_text)
```

```python
df['text_len'] = df['text'].apply(len)
```

```python
df.head()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['text_len'], bins=100)
```

```python
print(df.shape) # Before
df = df[ df['text_len'] <= 10_000 ]
print(df.shape) # After
```

```python
# Drop Duplicate
print(df.shape) # Before
df.drop_duplicates(['text'], inplace=True)
print(df.shape) # After
```

```python
df['datetime_stamp'] = df['date'].apply(lambda v: datetime.strptime(v, "%Y-%m-%d %H:%M:%S"))

df['datetime_stamp'] = df['datetime_stamp'].apply(lambda v: v.replace(hour=0, minute=0, second=0))
```

## Embedding


```python
model_id = "sentence-transformers/LaBSE"
embedding_model = SentenceTransformer(model_id)
```

```python
news_embeddings = embedding_model.encode(df['text'].values, show_progress_bar=True)
```

```python
news_embeddings.shape
```

# Topic Modeling Using BERTopic

# **dimensionality reduction**

```python
umap_model = UMAP(n_neighbors=15, n_components=15,
                  min_dist=0.0, metric='cosine',
                  random_state=101)
```

# **Clustering**

```python
# A higher min_cluster_size will generate fewer topics

hdbscan_model = HDBSCAN(min_cluster_size=50,
                        metric='euclidean',
                        cluster_selection_method='eom',
                        prediction_data=True)
```

## Vectorizer

```python
stop_words = stp.stopwords_list()

vectorizer_model = CountVectorizer(min_df=3,
                                   stop_words=stop_words,
                                   analyzer='word',
                                   max_df=0.5,
                                   ngram_range=(1, 3)
                                   )
```

# **Topic Representer**

```python

from bertopic.representation import KeyBERTInspired

keybert_model = KeyBERTInspired()

representation_model = {
    "KeyBERT": keybert_model
}
```

```python
from bertopic import BERTopic

topic_model = BERTopic(
    # Pipeline models
    embedding_model=embedding_model,
    umap_model=umap_model,
    hdbscan_model=hdbscan_model,
    vectorizer_model=vectorizer_model,
    representation_model=representation_model,

    # Hyperparameters
    top_n_words=10,
    verbose=True
)

topics, probs =  topic_model.fit_transform(
    df['text'].values,
    news_embeddings
)
```

```python
# Get the list of topics
topic_model.get_topic_info()
```

```python
# Visualize top topic keywords
topic_model.visualize_barchart(top_n_topics=12)
```

# **Topics Merging**

```python
topics_to_merge = [
    [15, 60, 4],
    [30, 23, 7]
]

topic_model.merge_topics(
    df['text'].values,
    topics_to_merge
)
```

```python
hierarchical_topics = topic_model.hierarchical_topics(
    df['text'].values
)

topic_model.visualize_hierarchy(
    hierarchical_topics=hierarchical_topics
)
```
